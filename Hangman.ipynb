```python
print(Hangman_logos_stages.logo)
print('''\nThis game is about guessing the word. If you guess correctly, then you can save the person from hanging.
Get ready!.''')


# Variables
num=-2
lives=6
word_list= Hangman_words.word_list

rep=0

# Use of words from a module containing random words.

chosen_word=word_list[random.randint(0,len(word_list)-1)]


display=['_' for i in chosen_word]

print(display)




while(display!=[*chosen_word]):
    if(lives<0):
        print('You lost')
        print(f'The choosen word was : {chosen_word}')
        break
    
    else:
    
        guessed_letter= input('Enter a letter:\n').lower()
        
        if(guessed_letter in display):
            print('This letter is already used. Use some other letter')
            
        if (guessed_letter in chosen_word):
            
            for i in range(len(chosen_word)):
                if chosen_word[i]==guessed_letter:
                    display[i]=guessed_letter
            
            print(display)
        else:
            num+=1             
            
            if(num<0):
                print(f'Your life left is {lives}.')
                continue
            else:
                
                print(stages_1[num])
                print(f'Your life left is {lives}.')
                lives-=1
            

            
        
        if(display==[*chosen_word]):
            print('You won')
            print(f'Hey, congratulations, you won. The word was {chosen_word}')
            break
       
```
